```python
# Generate summary using BERT-based model
summary = summarizer(text, max_length=150, min_length=50, do_sample=False)[0]['summary_text']
print("Generated Summary:")
print(summary)

```

```python
reference_summary = "The court found Rajesh Kumar guilty on all charges, including murder, robbery, and possession of stolen property. He was sentenced to life imprisonment for the murder charge under Section 302 of the IPC. He also received ten years of rigorous imprisonment for robbery under Section 397 and five years for receiving stolen property under Section 411."
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(reference_summary, summary)
```

```python
precision = scores['rouge1'].precision
recall = scores['rouge1'].recall
f1_score = scores['rouge1'].fmeasure

print(f"\nPrecision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```
