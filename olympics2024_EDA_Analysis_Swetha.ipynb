```python
medal_counts = data['medal_type'].value_counts()
medal_counts.plot(kind='bar', color=['brown', 'silver', 'gold'])
plt.title('Olympic Medals Distribution')
plt.xlabel('Medal Type')
plt.ylabel('Count')
plt.show()
```

```python
medal_counts = data['country_code'].value_counts()

# bar graph
medal_counts.plot(kind='bar', figsize=(14, 6))
plt.title('Olympic Medals Distribution based on Countries')
plt.xlabel('countries')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
country_counts = data['country_code'].value_counts()
top_countries = country_counts[country_counts > 10]
print(top_countries)
# bar graph
top_countries.plot(kind='bar', figsize=(14, 6))
plt.title('Countries with More Than 10 Olympic Medals')
plt.xlabel('Country')
plt.ylabel('Medal Count')
plt.show()
```

```python
gender_counts = data['gender'].value_counts()
gender_counts.plot(kind='bar', figsize=(8, 6))
plt.title('Total Number of Medals Won by Men and Women')
plt.xlabel('Gender')
plt.ylabel('Medal Count')
plt.show()
print(gender_counts)
```

```python
shooting_medals = data[data['discipline'] == 'Shooting']
country_counts = shooting_medals['country_code'].value_counts()
print("Number of medals won in Shooting by country:")
print(country_counts)
# bar graph
country_counts.plot(kind='bar', figsize=(14, 6))
plt.title('Shooting Medals Distribution based on Countries')
plt.xlabel('countries')
plt.ylabel('Count')
plt.show()
```

```python
discipline_counts = data['discipline'].value_counts()
plt.figure(figsize=(25, 25))
plt.pie(discipline_counts.values, labels=discipline_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Medals per Discipline')
plt.show()
```

```python
country_event_counts = data['country_code'].value_counts().reset_index()
country_event_counts.columns = ['country_code', 'event_count']
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world = world.merge(country_event_counts, left_on='iso_a3', right_on='country_code', how='left')
# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax)
world.plot(column='event_count', ax=ax, legend=True,
           legend_kwds={'label': "Number of Events Participated",
                        'orientation': "horizontal"},
           cmap='OrRd', missing_kwds={'color': 'lightgrey'})

plt.title('Countries Participation in Events')
plt.savefig('country_participation_choropleth.png', format='png', bbox_inches='tight')
plt.show()
```

```python
medal_counts = data.groupby(['medal_date', 'country_code']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
for country in medal_counts.columns:
    plt.plot(medal_counts.index, medal_counts[country], marker='o', label=country)

plt.title('Number of Medals Over Time by Country')
plt.xlabel('Date')
plt.ylabel('Number of Medals')
plt.legend(title='Country')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
